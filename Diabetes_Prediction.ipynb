Diabetes Prediction

Importing Dependencies

```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn import svm
from sklearn.metrics import accuracy_score
```

Data Collection

```python
# Loading the diabetes dataset into pandas dataframe
diabetes_dataset = pd.read_csv("diabetes.csv")
```

```python
diabetes_dataset
```

```python
# Displaying the first 5 rows of the dataset
diabetes_dataset.head()
```

```python
# Number of rows and columns of dataset
diabetes_dataset.shape
```

```python
# Getting the statistical measures of data
diabetes_dataset.describe()
```

```python
diabetes_dataset["Outcome"].value_counts()
```

1 -> Diabetic

0 -> Non-Diabetic

```python
diabetes_dataset.groupby("Outcome").mean()
```

```python
# Seperating the data
X = diabetes_dataset.drop(columns = "Outcome", axis = 1)
y = diabetes_dataset["Outcome"]
```

```python
print(X)
```

```python
print(y)
```

Splitting the Data

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y, random_state=2)
```

```python
scaler = StandardScaler()

X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

Training the Model

```python
classifier = svm.SVC(kernel="linear")
```

```python
# Training the support vector classifier
classifier.fit(X_train_scaled, y_train)
```

Model Evaluation

Accuracy Score

```python
# Accuracy score on Training data
X_train_prediction = classifier.predict(X_train_scaled)
acc_score = accuracy_score(X_train_prediction, y_train)

print(f"Accuracy score of the Training data: {acc_score}")
```

```python
# Accuracy score of Test data
X_test_prediction = classifier.predict(X_test_scaled)
acctest_score = accuracy_score(X_test_prediction, y_test)

print(f"Accuracy score of the Test data: {acctest_score}")
```

Making a Predictive System

```python
input_data = (5,116,74,0,0,25.6,0.201,30)

# Converting the input into a numpy array
np_array = np.asarray(input_data)

# Reshaping the data
input_data_reshaped = np_array.reshape(1, -1)

# Stanadardize the data
input_data_scaled = scaler.transform(input_data_reshaped)
print(input_data_scaled)

prediction = classifier.predict(input_data_scaled)
print(prediction)

if (prediction[0] == 1):
  print("Diabetic")
else:
  print("Non-Diabetic")
```
