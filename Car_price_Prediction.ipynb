Car Price Prediction

Importing the Dependencies

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import Lasso
from sklearn.linear_model import Ridge
from sklearn.linear_model import LinearRegression
from sklearn import metrics
```

Data Collection and Processing

```python
# # Loading the dataset
car_dataset = pd.read_csv("car data.csv")
```

```python
# Presenting the first 5 rows of the dataset
car_dataset.head()
```

```python
car_dataset.info()
```

```python
# Numbers of rows and columns
car_dataset.shape
```

```python
car_dataset.isnull().sum()
```

```python
# Checking the distribution of categorical data
print(car_dataset.Fuel_Type.value_counts())
print(car_dataset.Seller_Type.value_counts())
print(car_dataset.Transmission.value_counts())
```

Encoding the Categorical Data

```python
car_dataset.replace({"Fuel_Type": {"Petrol":0, "Diesel":1, "CNG":2}}, inplace=True)
car_dataset.replace({"Seller_Type": {"Dealer":0, "Individual":1}}, inplace=True)
car_dataset.replace({"Transmission": {"Manual":0, "Automatic":1}}, inplace=True)
```

```python
# To resolve downcasting problem

car_dataset["Fuel_Type"] = pd.to_numeric(car_dataset["Fuel_Type"])
car_dataset["Seller_Type"] = pd.to_numeric(car_dataset["Seller_Type"])
car_dataset["Transmission"] = pd.to_numeric(car_dataset["Transmission"])
```

```python
car_dataset.dtypes
```

```python
car_dataset.head()
```

Splitting the Data

```python
X = car_dataset.drop(columns = {"Car_Name", "Selling_Price"})
# Other way of dropping the columns - car_dataset.drop(["Car_Name", "Selling_Price"], axis=1)
y = car_dataset["Selling_Price"]
```

```python
print(X)
```

```python
print(y)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, stratify=y, random_state=2)
```

Training the Model

1. Linear Regression

```python
model = LinearRegression()
```

```python
model.fit(X_train, y_train)
```

Model Evaluation

```python
# Prediction on Training data
training_data_prediction = model.predict(X_train)
```

```python
# R squared error
error_score = metrics.r2_score(y_train, training_data_prediction)
print(f"R squared error: {error_score}")
```

Visualizing the Actual prices vs Predicted prices

```python
plt.scatter(y_train, training_data_prediction)
plt.xlabel("Actual Prices")
plt.xlabel("Predicted Prices")
plt.title("Actual vs Predicted prices")
plt.show()
```

```python
# Prediction on Test data
test_data_prediction = model.predict(X_test)
```

```python
# R squared error
error_score = metrics.r2_score(y_test, test_data_prediction)
print(f"R squared error: {error_score}")
```

```python
plt.scatter(y_test, test_data_prediction)
plt.xlabel("Actual Prices")
plt.xlabel("Predicted Prices")
plt.title("Actual vs Predicted prices")
plt.show()
```
